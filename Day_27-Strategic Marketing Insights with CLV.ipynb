```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/Superstore Datasets - segmentation - superstore.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df.info()
```

```python
df = df[['customer_id', 'order_id', 'order_date', 'sales', 'quantity']]
```

```python
df.head()
```

```python
df['amount'] = df['sales'] * df['quantity']
```

```python
df.head()
```

```python
df_rfm = df.groupby('customer_id').agg({'order_date':'max',
                               'order_id':'nunique',
                               'amount':'sum'}).reset_index()
```

```python
df_rfm
```

```python
max_date = df['order_date'].max()
max_date
```

```python
df_rfm['last_days_transactions'] = (max_date - df_rfm['order_date']).dt.days
df_rfm = df_rfm[['customer_id', 'last_days_transactions', 'order_id', 'amount']]
df_rfm.head()
```

```python
quartiles = df_rfm[['last_days_transactions', 'order_id', 'amount']].quantile(q=[0.25, 0.5, 0.75])
quartiles
```

```python
def r_score(x, c):
    if x <= c.loc[0.25]:
        return 1
    elif x <= c.loc[0.5]:
        return 2
    elif x <= c.loc[0.75]:
        return 3
    else:
        return 4
```

```python
def fm_score(x, c):
    if x <= c.loc[0.25]:
        return 4
    elif x <= c.loc[0.5]:
        return 3
    elif x <= c.loc[0.75]:
        return 2
    else:
        return 1
```

```python
df_rfm['R'] = df_rfm['last_days_transactions'].apply(r_score, args=(quartiles['last_days_transactions'],))
df_rfm['F'] = df_rfm['order_id'].apply(fm_score, args=(quartiles['order_id'],))
df_rfm['M'] = df_rfm['amount'].apply(fm_score, args=(quartiles['amount'],))

df_rfm.head()
```

```python
df_rfm['RFM'] = df_rfm['R'].astype(str) + df_rfm['F'].astype(str) + df_rfm['M'].astype(str)
df_rfm.head()
```

```python
df_rfm.groupby('M').agg({
    'amount':'median'
}).reset_index()
```

## Best Customers
Customers who BOught Most recently, most eften and spend the most.

### Marketing
No price incensive, New Product and loyalty programs

```python
# Filter the DataFrame for customers with RFM score '111'
Best_customer = df_rfm[(df_rfm['R']==1) & (df_rfm['F'] == 1) & (df_rfm['M'] == 1)]
print(Best_customer)
```

## Loyal Customers
Customers who bought most frequently

### Marketing
User R and M further segment

```python
import re

# Filter the DataFrame for customers with RFM score 'X1X'
loyal_customers = df_rfm.loc[(df_rfm['R'].apply(lambda x: bool(re.match(r'^[0-9]$', str(x)))) &
                               (df_rfm['F'] == 1) &
                               (df_rfm['M'].apply(lambda x: bool(re.match(r'^[0-9]$', str(x))))))]

if loyal_customers.empty:
    print("No loyal customers found.")
else:
    # Assign the label 'loyal customer' to these customers
    loyal_customers['Customer_Type'] = 'Loyal Customer'

    # View the resulting DataFrame
    print(loyal_customers)
```

## Big spenders
Customers who bought most receltly

### Marketing
Use F and M to further segment

```python
import re

# Filter the DataFrame for customers with RFM score 'XX1'
Big_spenders = df_rfm.loc[(df_rfm['R'].apply(lambda x: bool(re.match(r'^[0-9]$', str(x)))) &
                               (df_rfm['M'].apply(lambda x: bool(re.match(r'^[0-9]$', str(x)))))) &
                               (df_rfm['F'] == 1)]


if Big_spenders.empty:
    print("No Big Spenders found.")
else:
    # Assign the label 'Big Spenders' to these customers
    Big_spenders['Customer_Type'] = 'Big Spenders'

    # View the resulting DataFrame
    print(Big_spenders)
```
