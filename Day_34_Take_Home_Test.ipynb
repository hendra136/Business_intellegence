```python
quartiles = rfm[['recency', 'frequency', 'monetary']].quantile(q=[0.25, 0.5, 0.75])
quartiles
```

## Menghitung RFM

```python
def r_score(x, c):
    if x <= c.loc[0.25]:
        return 1
    elif x <= c.loc[0.5]:
        return 2
    elif x <= c.loc[0.75]:
        return 3
    else:
        return 4
```

```python
def fm_score(x, c):
    if x <= c.loc[0.25]:
        return 4
    elif x <= c.loc[0.5]:
        return 3
    elif x <= c.loc[0.75]:
        return 2
    else:
        return 1
```

```python
rfm['recency_score'] = rfm['recency'].apply(r_score, args=(quartiles['recency'],))
rfm['frequency_score'] = rfm['frequency'].apply(fm_score, args=(quartiles['frequency'],))
rfm['monetary_score'] = rfm['monetary'].apply(fm_score, args=(quartiles['monetary'],))
```

```python
rfm.head()
```

## RFM Segmentation

```python
rfm['RFM_SCORE'] = rfm['recency_score'].astype(str) + rfm['frequency_score'].astype(str) + rfm['monetary_score'].astype(str)
rfm.head()
```

```python
# Menambahkan kolom Segment_customer dengan ketentuan yang diberikan
def get_segment(rfm_score):
    if rfm_score == '111':
        return 'Best Customer'
    elif rfm_score == '331':
        return 'Almost Lost'
    elif rfm_score == '411':
        return 'Lost Customers'
    elif rfm_score == '444':
        return 'Lost Cheap Customers'
    elif len(rfm_score) == 3 and rfm_score[2] == '1' and rfm_score != '331' and rfm_score != '111' and rfm_score != '411':  # XX1
        return 'Big Spenders'
    elif len(rfm_score) == 3 and rfm_score[1] == '1' and rfm_score != '111' and rfm_score != '411':  # X1X
        return 'Loyal Customer'
    else:
        return 'Other'

# Menambahkan kolom Segm_customer
rfm['Segment_customer'] = rfm['RFM_SCORE'].apply(get_segment)

# Menampilkan DataFrame hasil
rfm
```

## 2. Menghitung ekspektasi nilai yang dihasilkan untuk setiap konsumen untuk 3 bulan ke depan menggunakan algoritma Gamma-Gamma

## CLTV

```python
df_cleaned.head()
```

```python
T = (df_cleaned['date'].max() - df_cleaned.groupby('customer_id')['date'].min()).dt.days
```

```python
# Mempersiapkan data untuk visualisasi BG-NBD model
frequency = df_cleaned.groupby('customer_id').size()
recency = df_cleaned.groupby('customer_id')['date'].apply(lambda x: (x.max() - x.min()).days)
T = (df_cleaned['date'].max() - df_cleaned.groupby('customer_id')['date'].min()).dt.days
```

```python
# Membuat penjumlahan data frame
summary = pd.DataFrame({
    'customer_id': frequency.index,
    'frequency': frequency.values,
    'recency': recency.values,
    'T': T.values
})
```

```python
# Install Modul
!pip install lifetimes

from lifetimes import BetaGeoFitter, GammaGammaFitter
```

```python
# Fit the BG-NBD model
bgf = BetaGeoFitter(penalizer_coef=0.0)
bgf.fit(summary['frequency'], summary['recency'], summary['T'])
```

```python
# Memprediksi lifetime pelanggan selama 3 bulan kedepan
summary['predicted_purchases'] = bgf.conditional_expected_number_of_purchases_up_to_time(3*30, summary['frequency'], summary['recency'], summary['T'])
```

### **Ekspektasi nilai yang dihasilkan untuk setiap konsumen untuk 3 bulan ke depan**

```python
# Persiapan Gamma-Gamma model
monetary = df_cleaned.groupby('customer_id')['total'].mean()
```

```python
# Fit the Gamma-Gamma model
ggf = GammaGammaFitter(penalizer_coef=0.0)
ggf.fit(summary['frequency'], monetary)
```

```python
# Memprediksi lifetime pelanggan selama 3 bulan ke depan
summary['monetary'] = monetary.values
summary['predicted_clv'] = ggf.customer_lifetime_value(
    bgf, summary['frequency'], summary['recency'], summary['T'], summary['monetary'], time=3*30
)
```
